```python
import datetime
import platform
import time

from dp.env import DroneRescueEnv
from dp.mdp import value_iteration
from dp.visualization import plot_policy, plot_value_heatmap
from dp.analysis import summarize_dp_results
from parameters import GROUP_NUMBER

GROUP_ID = GROUP_NUMBER  # Team number imported from shared parameters
timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
vm_id = platform.node()
print(f'Execution Timestamp: {timestamp}')
print(f'Virtual Machine ID: {vm_id}')
print(f'Group ID: {GROUP_ID}')
```

## 1. Drone Environment + State / Action Design

Instantiate the custom environment and describe the state representation.

```python
env = DroneRescueEnv(GROUP_ID)
print('Grid size:', env.grid_size)
print('Max battery:', env.max_battery)
print('Wind probability:', env.wind_prob)
print('Max steps:', env.max_steps)
print('Rescue positions:', env.rescue_positions)
print('Charging positions:', env.charging_positions)
print('Danger positions:', env.danger_positions)
print('Blocked positions:', env.blocked_positions)
```

## 2. Value Iteration

Enumerate reachable states and compute the optimal policy.

```python
reachable_states = env.enumerate_reachable_states()
print('Reachable states:', len(reachable_states))

start_time = time.time()
value, policy, iterations, delta = value_iteration(env, reachable_states)
runtime = time.time() - start_time
metadata = {
    'iterations': iterations,
    'runtime': runtime,
    'delta': delta,
    'start_value': value[reachable_states.index((0, 0, env.max_battery, (1 << len(env.rescue_positions)) - 1, 0))],
}
print(summarize_dp_results(metadata))
```

## 3. Visualization

Plot the learned policy and the state-value heatmap for a meaningful slice.

```python
state_to_index = {state: idx for idx, state in enumerate(reachable_states)}
battery_full = env.max_battery
rescue_mask = (1 << len(env.rescue_positions)) - 1
plot_policy(env, policy, battery_full, rescue_mask)
plot_value_heatmap(env, value, state_to_index, battery_full, rescue_mask)
```
