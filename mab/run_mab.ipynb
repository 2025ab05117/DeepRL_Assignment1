```python
import datetime
import platform

from mab.dataset import create_patient_dataset
from mab.environment import MABEnvironment
from mab.strategies import greedy_strategy, epsilon_greedy_strategy, ucb1_strategy
from mab.analysis import plot_cumulative_rewards, summarize_results
from parameters import GROUP_NUMBER

timestamp = datetime.datetime.now().strftime('%Y-%m-%d %H:%M:%S')
vm_id = platform.node()
print(f'Execution Timestamp: {timestamp}')
print(f'Virtual Machine ID: {vm_id}')
print(f'Group Number: {GROUP_NUMBER}')
```

## 1. Dataset + Environment Setup

Generate the synthetic patient dataset and instantiate a reusable MAB environment.

```python
dataset, metadata = create_patient_dataset(GROUP_NUMBER)
print('Group number:', metadata['group_number'])
print('Number of medicines:', metadata['n_medicines'])
print('Hidden success probabilities:', metadata['hidden_probs'])
dataset.head(10)
```

```python
env = MABEnvironment(metadata['hidden_probs'], dataset)
print('Environment initialized with', env.n_arms, 'arms')
```

## 2. Greedy + Epsilon-Greedy Strategies

Run the immediate exploitation policy and epsilon-greedy variants.

```python
_, greedy_history = greedy_strategy(env)
_, egreedy_10_history = epsilon_greedy_strategy(env, epsilon=0.10)
_, egreedy_1_history = epsilon_greedy_strategy(env, epsilon=0.01)
_, egreedy_50_history = epsilon_greedy_strategy(env, epsilon=0.50)
```

## 3. UCB1 + Comparative Analysis

Run UCB1 and compare all strategies on cumulative reward.

```python
_, ucb_history = ucb1_strategy(env)
history_map = {
    'Immediate Exploitation': greedy_history['cumulative_reward'],
    'Epsilon-Greedy 10%': egreedy_10_history['cumulative_reward'],
    'Epsilon-Greedy 1%': egreedy_1_history['cumulative_reward'],
    'Epsilon-Greedy 50%': egreedy_50_history['cumulative_reward'],
    'UCB1': ucb_history['cumulative_reward'],
}
results = summarize_results(history_map)
print('Final cumulative rewards:')
for name, value in results.items():
    print(f'  {name}: {value:.2f}')
```

```python
plot_cumulative_rewards(history_map)
```
